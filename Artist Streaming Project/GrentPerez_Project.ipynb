Section 3 - Vizualizations on my dataset

```python
df["Release Date"] = pd.to_datetime(df["Release Date"])
df = df.sort_values("Release Date")
```

```python
plt.figure(figsize=(12,6))
plt.scatter(df["Release Date"], df["Streams"] / 1_000_000)
plt.title("Streams by Release Date")
plt.xlabel("Release Date")
plt.ylabel("Streams (Millions)")
plt.xticks(rotation=45)
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()
```
